## Matplotlib
- Categorical 
    - Bar Chart
    - Line Chart
- Numerical
    - Scatter Plot

```python
# importing libraries
import matplotlib.pyplot as plt
import numpy as np
import math
```

### Bar Chart

```python
animals = ['Giraffe', 'Crocodile', 'Elephant']
height = [7, 6, 3]
age = [25, 50, 80]
```

```python
plt.bar(animals, height)
```

### Line Chart

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

### Scatter Plot

```python
plt.scatter(height, age)
plt.suptitle('Height: x, Age: y')
```

```python
plt.scatter(age, height)
plt.suptitle('Age: x, Height: y')
```

## Combine Charts

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First')

axes[1].plot(y,x)
axes[1].set_title('Second')
```
